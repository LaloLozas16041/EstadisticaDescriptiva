### UNO

##### **Problema:** Dado un conjunto de datos con información de clientes de un banco, identifica qué variables influyen más en la pérdida de clientes y propone estrategias para reducir la tasa de churn.

### 

- ClientID: ID of the client.
- Age: Age of the client.
- Account_Balance: Money in the client's bank account.
- Number_of_Products: Number of banking products the client uses (loans, credit cards, etc.).
- Credit_Score: Client's credit score.
- IsActiveMember: Whether the client is active or not (1 for Yes, 0 for No).
- EstimatedSalary: Estimated annual salary of the client.
-  Country: Country of the client.
- Exited: Whether the client left the bank or not (1 for Yes, 0 for No).

```python
import pandas as pd
import numpy as np

np.random.seed(42) 

data_size = 10000

df = pd.DataFrame({
    'ClienteID': np.arange(1, data_size+1),
    'Edad': np.random.randint(18, 70, size=data_size),
    'Genero': np.random.choice(['Masculino', 'Femenino'], size=data_size),
    'Saldo': np.random.uniform(0, 100000, size=data_size),
    'NumProductos': np.random.randint(1, 5, size=data_size),
    'TarjetaCredito': np.random.choice([0, 1], size=data_size),
    'Activo': np.random.choice([0, 1], size=data_size),
    'ScoreCredito': np.random.randint(300, 850, size=data_size),
    'Churn': np.random.choice([0, 1], size=data_size)
})

indices_nulos = np.random.randint(0, data_size, 500)
df.loc[indices_nulos, 'Saldo'] = np.nan

indices_nulos = np.random.randint(0, data_size, 300)
df.loc[indices_nulos, 'ScoreCredito'] = np.nan
```

```python
df
```

```python
df.to_csv('dfUNO.cdv')
```

### OCHO

### 

```python
import numpy as np
import pandas as pd

np.random.seed(42)

data_size = 10000
df_pricing = pd.DataFrame({
    'Distance_km': np.random.uniform(1, 50, data_size),
    'Time_of_day': np.random.choice(['Morning', 'Afternoon', 'Evening', 'Night'], data_size),
    'Day_of_week': np.random.choice(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], data_size),
    'Demand': np.random.choice(['Low', 'Medium', 'High'], data_size),
    'Event': np.random.choice([0, 1], data_size), # 0: No event, 1: There's an event
})

# Simulate price based on the factors
base_price = 10
df_pricing['Price'] = (base_price + 
                      df_pricing['Distance_km'] * 0.5 + 
                      df_pricing['Time_of_day'].replace({'Morning': 1, 'Afternoon': 0, 'Evening': 2, 'Night': 3}) + 
                      df_pricing['Demand'].replace({'Low': -2, 'Medium': 0, 'High': 3}) + 
                      df_pricing['Event'] * 5 + 
                      np.random.normal(0, 2, data_size))

# Introducing missing values
indices_null = np.random.randint(0, data_size, 500)
df_pricing.loc[indices_null, 'Price'] = np.nan
```
